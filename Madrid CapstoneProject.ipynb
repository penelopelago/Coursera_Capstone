### Fifth Part: Bend our data for the analysis

A) CREATE A ONEHOT DATAFRAME

Since we are working with a variable that contains categories, we will be using onehot: we will create a dataframe where every venue category takes the value of 1 or 0, depending on whether that venue exist or nor for an specific neighborhood.

With get_dummies we can choose one column and every value of the column will become a header. So instead of having one column called Clear_category, we will obtain as many  columns as unique venues.

```python

madrid_onehot = pd.get_dummies(df_definitive[['Clear_category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
madrid_onehot['neighborhood_name'] = df_definitive['neighborhood_name'] 

# move neighborhood column to the first column
fixed_columns = [madrid_onehot.columns[-1]] + list(madrid_onehot.columns[:-1])
madrid_onehot = madrid_onehot[fixed_columns]

madrid_onehot.head()
```

B) OBTAINIG A DATA FRAME WITH THE FRECUENCY FOR EACH VENUE

Next, let's group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
madrid_grouped = madrid_onehot.groupby('neighborhood_name').mean().reset_index()
madrid_grouped
```

C) CREATE A FUNCTION WHICH RETURN THE MOST COMMON VENUES IN DESCENDING ORDER

```python
#First, let's write a function to sort the venues in descending order.

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#Now let's create the new dataframe and display the top 10 venues for each neighborhood

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['neighborhood_name']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['neighborhood_name'] = madrid_grouped['neighborhood_name']

for ind in np.arange(madrid_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(madrid_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Sixth Part: Clusters

A) IMPORTING TOOLS


```python
# import k-means from clustering stage
from sklearn.cluster import KMeans

#We are going to visualize our results in a map we need the following tools
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

B) GROUP NEIGHBORHOODS INTO CLUSTERS

We aim to obtain a Dataframe that includes a column with the clusters

```python
#Run k-means to cluster the neighborhood into 5 clusters.

# set number of clusters
kclusters = 5

madrid_grouped_clustering = madrid_grouped.drop('neighborhood_name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(madrid_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
#Let's create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

print(type(neighborhoods_venues_sorted))
neighborhoods_venues_sorted.tail()
```

```python
#If I join two dataframes but there are missing values, the Cluster Label column will change from integer to float and that would make impossible our future analysis.
#To avoid that I will have to eliminate the district 7 from the analysis
madrid_merged = df_data.drop([7], axis=0) 

# merge madrid_grouped with df_data to add latitude/longitude for each neighborhood
madrid_merged = madrid_merged.join(neighborhoods_venues_sorted.set_index('neighborhood_name'), on='Name')

madrid_merged.head()
```

C) VISUALIZATION

It will help us to see the clusters in a map. Every neighborhood will appear with the color of its cluster

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(madrid_merged['Latitude'], madrid_merged['Longitude'], madrid_merged['Name'], madrid_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
